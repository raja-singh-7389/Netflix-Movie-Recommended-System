```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import joblib

# Step 1: Load and preprocess the dataset
def load_and_preprocess_data():
    # Load the Netflix titles dataset
    movie = pd.read_csv('/content/Netflix-title.csv.csv', on_bad_lines='skip')

    # Keep only relevant columns
    movie = movie[['Movie Name', 'Genre']]

    # Drop rows with missing values
    movie.dropna(inplace=True)

    # Convert movie names and genres to lowercase for consistency
    movie['Movie Name'] = movie['Movie Name'].str.lower()
    movie['Genre'] = movie['Genre'].str.lower()

    return movie

# Step 2: Create similarity matrix
def build_similarity_matrix(movie):
    # Use CountVectorizer to create a Bag of Words representation of genres
    cv = CountVectorizer(max_features=5000, stop_words='english')
    vector = cv.fit_transform(movie['Genre']).toarray()

    # Compute cosine similarity matrix
    similarity = cosine_similarity(vector)

    return similarity

# Step 3: Save the processed data to joblib files
def save_joblib_files(movie, similarity):
    # Save the movie DataFrame as a joblib file
    joblib.dump(movie, 'movie_list.pkl')

    # Save the similarity matrix as a joblib file
    joblib.dump(similarity, 'similarity.joblib')

    print("Joblib files for 'movie_list.pkl' and 'similarity.joblib' have been created successfully.")

# Step 4: Define the recommendation function
def recommend(movie_name):
    # Load the data and similarity matrix
    movie = joblib.load('movie_list.pkl')
    similarity = joblib.load('similarity.joblib')

    # Convert the input movie_name to lowercase for consistency
    movie_name = movie_name.lower()

    # Check if the movie exists in the dataset
    if movie_name not in movie['Movie Name'].values:
        print(f"Movie '{movie_name}' not found in the database.")
        return

    # Get the index of the movie
    movie_index = movie[movie['Movie Name'] == movie_name].index[0]

    # Get similarity scores for the selected movie
    distances = similarity[movie_index]

    # Get indices of the top 5 most similar movies (excluding itself)
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # Print the recommended movies
    print(f"Top 5 movies similar to '{movie_name}':")
    for i in movie_list:
        print(movie.iloc[i[0]]['Movie Name'])

# Step 5: Run the process
movie = load_and_preprocess_data()
similarity = build_similarity_matrix(movie)
save_joblib_files(movie, similarity)

# Example usage of the recommendation function
recommend('ganglands')  # Replace with any movie name you want to recommend
```
